```python
import pandas as pd
import random
```

```python
temp = ['Low', 'Mild', 'High']
wind = temp
weather = ['Sunny', 'Rainy', 'Overcast']
courtCon = ['Grassy', 'Clay', 'Cemented']
court = ['Open Air', 'Enclosed Environment']
humidity = temp
```

```python
# for training
df = pd.DataFrame()
# for testing
df2 = pd.DataFrame()
```

```python
def willPlay(dic):
    if (dic['Temperature'] == 'Low' and dic['Wind'] =='High') or dic['Humidity'] == 'High':
        return 'No'
    if dic['Weather'] == 'Rainy' and (not dic['Court'] == 'Enclosed Environment'):
        return 'No'
    if dic['Weather'] == 'OverCast' and dic['Court Condition'] == 'Clay':
        return 'No'
    return 'Yes'
```

```python
def addRow(df):
    dic = {}
    dic['Temperature'] = random.choice(temp)
    dic['Wind'] = random.choice(wind)
    dic['Weather'] = random.choice(weather)
    dic['Court Condition'] = random.choice(courtCon)
    dic['Court'] = random.choice(court)
    dic['Humidity'] = random.choice(humidity)
    dic['Will Play'] = willPlay(dic)
    newdf = df.append(dic, ignore_index=True)
    return newdf
```

```python
# for training
for i in range(400):
    df = addRow(df)
# for testing
for i in range(20):
    df2 = addRow(df2)
```
